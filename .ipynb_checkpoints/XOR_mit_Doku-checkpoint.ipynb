Sigmoid function <br>
$\sigma(x) = \frac{1}{1 + e^{-x}}$ = y <br>
$\sigma'(x) = \sigma(x) \cdot (1-\sigma(x))$ <br>
Da die Sigmoidfunktion bereits berechnte wurde lautet die Ableitung nur noch: <br>
$\sigma'(x) = y \cdot (1-y)$ 

Raval Activation-Functions: <br>
https://github.com/llSourcell/Which-Activation-Function-Should-I-Use/blob/master/activations.py

```python
def sigmoid(x, deriv = False):
    if(deriv == True):
        return x * (1-x)
    return 1 / (1+np.exp(-x))
```

```python
def tanh(x, deriv = False):
    if(deriv == True):
        return 1 - x**2
    return np.tanh(x)        
```

```python
def relu(x, deriv = False):
    if(deriv == True):
        x[x > 0] = 1
        x[x < 0] = 0
    x[x < 0] = 0
    return x
```

```python
def weight_init(layer1, layer2, typ = 'He'):
    
    # where randn samples FROM a zero mean, unit standard deviation gaussian
    w_rand = np.random.randn(layer2, layer1)
    
    # Lecture 6-48
    if(typ == 'Xavier'):
        return w_rand * np.sqrt(1 / layer2)
    
    # Wird manchmal auch eingesetzt von
    # Andrew ng (https://www.youtube.com/watch?v=s2coXdufOzE)
    elif(typ == 'Xavier_1'):
        return w_rand * sqrt(2 / (layer1 +layer2))

    # Lecture 6-51 ours
    # for use in practice in the specific case of neural networks with ReLU neurons (http://cs231n.github.io/neural-networks-2/)
    # Andrew ng (https://www.youtube.com/watch?v=s2coXdufOzE)
    elif(typ == 'He'):
        return w_rand * np.sqrt(2 / layer2)
```

```python
def svm_loss(w, x, delta, y):
    # loss berechnen
    num_train = x.shape[1]
    scores = np.dot(w, x)
    margins = np.maximum(0, scores - scores[y, np.arange(num_train)] + delta)
    margins[y, np.arange(num_train)] = 0
    loss = np.sum(margins) / num_train
    
    # dW berechnen
    x_mask = np.zeros(margins.shape)
    x_mask = x_mask[margins > 0] = 1
    incorrect_counts = np.sum(margins, axis = 0)
    x_mask(y, np.arange(num_train)) = -incorrect_counts
    dW = np.dot(x_mask, x.T) / num_train
    
    return loss, dW
```

```python
def softmax_loss(w, x, y):
    # loss berechnen
    num_train = x.shape[1]
    scores = np.dot(w, x)
    unnormalized_margins = np.exp(scores[y, np.arange(num_train)]) / np.sum(scores, axis = 0)
    margins = -np.log(unnormalized_margins)
    loss = np.sum(margins) / num_train
    
    # dW berechnen
    x_mask = margins
    x_mask[y, np.arange(num_rain)] -= 1 
    dW = np.dot(x_mask, x.T) / num_train
    
    return loss, dW
```
